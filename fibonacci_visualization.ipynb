```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Upload the CSV file
uploaded = files.upload()

# Load the CSV into a DataFrame
data = pd.read_csv(next(iter(uploaded.keys())))

# Columns for plotting
metrics = ['sim_insts', 'sim_ticks', 'Miss Rate']
cpu_configs = data['CPU'] + " (" + data['Memory Config'] + ")"

# Generate bar charts for each metric
for metric in metrics:
    plt.figure(figsize=(12, 6))
    plt.bar(cpu_configs, data[metric], color='skyblue', alpha=0.8)
    plt.title(f"{metric} vs CPU Configurations")
    plt.xlabel("CPU Configurations")
    plt.ylabel(metric)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
